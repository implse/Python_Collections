# Counter

A counter is a container that stores elements as dictionary keys, and their counts are stored as dictionary values.
Counter is a subclass of dict. Therefore it is an unordered collection where elements and their respective count are stored as dictionary.

```python
from collections import Counter

# List of students
class_1 = ["Bob", "Becky", "Chad", "Darcy", "Frank", "Hannah", "Kevin", "Harry", "Melanie", "Harry", "Penny", "Steve"]

# Create a counter for class 1
c1 = Counter(class_1)
print(c1)

# How many students name Harry
print(c1["Harry"])

# How many students in class 1
print(sum(c1.values()), "students in class 1.")
```

```python
class_2 = ["Harry", "Gary", "Oliver", "John", "Mary", "Justin", "Molly"]

# Create a counter for class 1
c2 = Counter(class_2)

# Combine two class
c1.update(class_2)
print(c1)

# What's the most common name in the two classes
print(c1.most_common(2))

# What's common between two classes
print(c1 & c2)
```
